```python
# Loop para processar até 20 imagens
contador = 0
while contador < 20:
    # Caminho local da imagem de exemplo (substitua pelo caminho real do arquivo)
    img_path = '/content/imagemTexto/imagem{}.jpg'.format(contador+1)

    # Abrir a imagem
    img = Image.open(img_path)

    # Executar OCR na imagem
    texto = pytesseract.image_to_string(img, lang='por')

    # Exibir o texto extraído
    print("Texto extraído da imagem {}: \n{}".format(contador + 1, texto))

    # Incrementar o contador
    contador += 1
```
